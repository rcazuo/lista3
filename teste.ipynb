```python
from control import *
from control.matlab import *
import matplotlib.pyplot as plot
from sympy import symbols

a = [3,3]
b = [1,4,6,4]
g = tf(a,b)

print (g)
```

```python
polo1 = g.pole()
zero1 = g.zero()
ganho1 = g.dcgain()

print(polo1)
print(zero1)
print(ganho1)
```

```python
[polo2,zero2,ganho2] = tf2zpk(b,a)
print(polo2)
print(zero2)
print(ganho2)
```

```python
pzmap(g,plot=True,grid=True,title="Mapeamento de polos e zeros")
```

```python
[wn,amortecimento,polos]=damp(g,False)
print(wn)
print(amortecimento)
print(polos)
```
